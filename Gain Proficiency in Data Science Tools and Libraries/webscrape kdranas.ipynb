````python
# <Response [200]> means success code
# using try catch to avoid errors or jam

try:
    # making a request with GET
    r = requests.get("https://nkiri.com")
    # parsing the Html
    soup = bs(r.content, 'html.parser')
    print(soup.prettify())
except Exception as e:
    print(f"Error:", e)
````
