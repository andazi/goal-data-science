How to decide if the data point should be included to in the dataset for further analysis?

#### we use confidence level with quantile() / percent

```python
setosa_df = df[df['species']=='Iris-setosa']
```

```python
setosa_df['petal_length'].quantile(0.97) # 97%
```

This means 97% of the data point would be determine by the above value

```python
setosa_df['petal_length'].quantile(0.03) # 3%
```

```python
setosa_df['petal_length'].describe(percentiles=[0.03,0.97])
```

```python
df[(df['species']=='Iris-setosa') & (df['petal_length'] > setosa_df['petal_length'].quantile(0.97)) | (df['petal_length'] < setosa_df['petal_length'].quantile(0.03))]
```

The above are possible outliers and needs to drop due to how few it is
