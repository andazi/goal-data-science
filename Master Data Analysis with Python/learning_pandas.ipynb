**pat (pattern):** The string or regular expression pattern to search for in each element of the Series.

**case:** Boolean parameter for a case-sensitive search (default is True). Set to False for case-insensitive search.

**flags:** Optional parameter for modifying behavior with regular expression patterns, e.g., using re.IGNORECASE for case-insensitive search.

**na:** Treatment of missing values (NaN). Default is "no," treating NaN as False. Set to "False" or "True" to explicitly include or exclude NaN values.


**regex:** Boolean parameter indicating if pat should be treated as a regular expression. False treats pat as a literal substring, while True interprets it as a regular expression.
